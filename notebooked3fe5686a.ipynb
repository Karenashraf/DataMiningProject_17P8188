Input data that will be processed

```python
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
train_data.head()
```

Read train data

```python
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")
test_data.head()
```

Read data to be tested

```python
c = ["Name","Ticket","Cabin", "PassengerId"]
train_data = train_data.drop(c, axis=1)
train_data.info()
```

Drop 4 columns that contains name, ticket, cabin and ID information since these data will not matter with the final predictions, we can't get any useful conclusion from them

```python
train_data.head()
```

```python
train_data['Age'] = train_data['Age'].interpolate()
train_data['Fare'] = train_data['Fare'].interpolate()
train_data ['Age'] = np.float32(train_data['Age'])
train_data['Embarked'].fillna("Missing", inplace=True)

train_data.info()
train_data.head()
```

Using interpolatio  to fill in missing values in age and fare columns, and filling missing data in embarked column with the word "missing"

```python
from sklearn import preprocessing
train_data['Sex'] = preprocessing.LabelEncoder().fit_transform(train_data['Sex'])
train_data['Embarked'] = preprocessing.LabelEncoder().fit_transform(train_data['Embarked'])
train_data[['Sex','Embarked']].info()

print(train_data)

train_data ['Survived'] = np.float32(train_data['Survived'])
train_data ['Sex'] = np.float32(train_data['Sex'])
train_data ['Pclass'] = np.float32(train_data['Pclass'])
train_data ['Embarked'] = np.float32(train_data['Embarked'])
train_data ['Fare'] = np.float32(train_data['Fare'])
train_data ['Parch'] = np.float32(train_data['Parch'])
train_data ['SibSp'] = np.float32(train_data['SibSp'])
train_data.info()
```

Sex and Embarked columns are filled with text data not numbers so they are converted to numbers that represent the textual information. Then all columns are casted to float32 data type

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

i = train_data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']]
j = train_data['Survived']
mdl = DecisionTreeClassifier(random_state=42)
scores = cross_val_score(mdl, i, j, cv=5)
print(scores)
mdl.fit(i,j)
```

i variable contains all columns except survived and j variable contains the survived column. Decision Tree model is created with random state 42 and then cross validation is performed.

```python
PD=test_data['PassengerId']
cols1 = ["Name","Ticket","Cabin", "PassengerId"]
test_data = test_data.drop(cols1, axis=1)
test_data['Embarked'].fillna("Missing", inplace=True)
test_data['Age'] = test_data['Age'].interpolate()
test_data['Fare'] = test_data['Fare'].interpolate()
test_data['Sex'] = preprocessing.LabelEncoder().fit_transform(test_data['Sex'])
test_data['Embarked'] = preprocessing.LabelEncoder().fit_transform(test_data['Embarked'])
test_data ['Sex'] = np.float32(test_data['Sex'])
test_data ['Pclass'] = np.float32(test_data['Pclass'])
test_data ['Embarked'] = np.float32(test_data['Embarked'])
test_data ['Fare'] = np.float32(test_data['Fare'])
test_data ['Parch'] = np.float32(test_data['Parch'])
test_data ['SibSp'] = np.float32(test_data['SibSp'])
test_data ['Age'] = np.float32(test_data['Age'])

test_data.info()
test_data.head(100)
```

Same data cleaning that was performed on train_data is repeated for test_data

```python
predictions = mdl.predict(test_data)
sub = pd.DataFrame({'PassengerId':PD, 'Survived':predictions})
sub.Survived = sub.Survived.astype(int)
sub.to_csv( './sub.csv',index=False )
```
